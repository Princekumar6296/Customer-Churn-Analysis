```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 

df = pd.read_csv('customer churn data.csv')
df.head()
```

#Replacing blanks with 0 as tenure is 0 and no total charges are recorded beacuse they join the company few monts ago

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype('float')
```

```python
def vl(value):
    if value == 1:
        return  "yes"
    else:
        return  " no"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(vl)
```

#Converted 0 and 1 values of SeniorCitizen to Yes/No to make easier to understand 

```python
Pg= sns.countplot(x = 'Churn', data = df)

Pg.bar_label(Pg.containers[0])
plt.title("count of customer")
plt.show()
```

```python
plt.figure(figsize = (4,4))
gh = df.groupby('Churn').agg({'Churn':"count"})

plt.pie(gh['Churn'], labels = gh.index, autopct="%1.2f%%")
plt.title("Percentage of churn data",fontsize=10)
plt.show()
```

```python
plt.figure(figsize = (4,4))
gh = sns.countplot(x = "gender", data = df, hue = "Churn")

for container in gh.containers:
    gh.bar_label(container)
plt.show()
```

#The numbers of employee working with us and left the company.

```python
plt.figure(figsize = (4,4))
ab = sns.countplot(x = 'SeniorCitizen' , data = df)

ab.bar_label(ab.containers[0])
plt.show()
```

```python
plt.figure(figsize = (4,4))
counts = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# 2. Convert counts to percentages (row-wise)
percentages = counts.div(counts.sum(axis=1), axis=0) * 100

# 3. Plot the stacked bar chart
ax = percentages.plot(kind='bar', stacked=True, figsize=(6,5), colormap='viridis') #for changing the color color=['#1f77b4', '#ff7f0e']

# 4. Add percentage labels
for i, row in enumerate(percentages.values):
    bottom = 0
    for j, val in enumerate(row):
        if val > 0:
            ax.text(i, bottom + val/2, f'{val:.1f}%', ha='center', va='center', color='white', fontsize=10)
            bottom += val

# 5. Final touches
plt.title('Churn Rate by Senior Citizen Status (as %)')
plt.xlabel('Senior Citizen')
plt.ylabel('Percentage')
plt.legend(title='Churn')
plt.xticks(ticks=[0,1], labels=['Not Senior', 'Senior'])
plt.tight_layout()
plt.show()
```

#The plot shows that **churn is highest among customers with very short tenure**, especially in the first few months. As tenure increases, the churn rate drops significantly, indicating that **long-term customers are much more likely to stay**.

```python
df['SeniorCitizen'] = df['SeniorCitizen'].astype(str)

# Categorical columns
categorical_cols = [
    'gender', 'SeniorCitizen', 'Partner', 'Dependents',
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies',
    'Contract', 'PaperlessBilling', 'PaymentMethod', 'Churn'
]

# Plot setup
n_cols = 3
n_rows = -(-len(categorical_cols) // n_cols)  # Ceiling division

plt.figure(figsize=(20, n_rows * 4))

for i, col in enumerate(categorical_cols, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.countplot(data=df, x=col, hue = 'Churn', palette='pastel', order=df[col].value_counts().index)
    plt.title(f'{col}')
    plt.xticks(rotation=45)
    # plt.legend('false')

plt.tight_layout()
plt.show()
```

#Most customers did not churn, but certain features show a clear pattern—month-to-month contracts, electronic checks, and lack of services like tech support and online security are associated with higher churn. Features like longer-term contracts and having partners or dependents appear linked with customer retention. Overall, categorical service features show noticeable differences in churn behavior across their categories.
